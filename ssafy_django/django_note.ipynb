django project 생성 순서

1. working directory 만들기
2. 가상 환경 설정 (python -m venv [venv 이름])
3. 설정한 가상 환경 실행 .source/Script/activate
4. pip list (가상환경 내 설치된 목록보기)
5. pip install django==3.2.18
6. django-admin startproject [pjt 이름]
7. python manage.py runserver 로 서버 실행

> 127.0.0.1:8000 (localhost) - loopback : 주소 외부로 나가지 않음 <br>
> 127.0.0.1 : 컴퓨터 주소, 8000 : 포트번호

- 각기 다른 버전의 라이브러리, 프레임워크등을 사용하기 위해 그 프로젝트 구성을 꾸리기 위해 사용

> python -m venv [venv 이름] : virture environment module 을 실행하여 가상환경 생성 <br>
> venv/Scripts/activate => activate : bash, activate.bat : batch, activate.ps1 : powershell

<br>

- django 를 구동시키기 위해 최소 1개 이상의 app 생성이 필요

## Day 2

- Django 에 적용된 디자인 패턴은 MTV 패턴
- MTV 패턴은 MVC 디자인 패턴을 기반으로 조금 변형된 패턴

### MVC 패턴이란 ?

1. Model(Data) - View(화면) - Controller(중개자역할) 의 줄임말
2. 데이터 및 논리 제어를 구현하는데 널리 사용되는 소프트웨어 디자인 패턴
3. 하나의 큰 프로그램을 세가지 역할로 구분한 개발 방법론

- MVC 소프트웨어 디자인 패턴의 목적 : 각 부분을 독립적으로 개발하는데 용이

<br>

### MTV 패턴이란?

1. Model(Data) - Template(화면) - View(중개자역할) 의 줄임말
2. MVC 패턴과 매우 유사하지만 처리하는 각 시스템의 이름이 조금 다름

- Model (Data) - 데이터베이스 관리
- Template(화면) - 보여지는 HTML 관리
- View(중개자역할) - Model & Template 과 관련한 로직을 처리해서 응답을 반환
